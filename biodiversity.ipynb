```python
import pandas as pd
import numpy as np
import seaborn
from matplotlib import pyplot as plt 


species = pd.read_csv('species_info.csv')
observations = pd.read_csv('observations.csv')

print(species.head(10))
print(observations.head(10))
print(np.unique(species.category))
print(np.unique(observations.park_name))
```

```python
amphibians = species[(species.category == 'Amphibian')]
birds = species[(species.category == 'Bird')]
fish = species[(species.category == 'Fish')]
mammals = species[(species.category == 'Mammal')]
nonvascular_plants = species[(species.category == 'Nonvascular Plant')]
reptiles = species[(species.category == 'Reptile')]
vascular_plants = species[(species.category == 'Vascular Plant')]

recovering = species[(species.conservation_status == 'In Recovery')]
threatened = species[(species.conservation_status == 'Threatened')]
concern = species[(species.conservation_status == 'Species of Concern')]
endangered = species[(species.conservation_status == 'Endangered')]
```

```python
#Examination of categories and conservation status
protected_animals = species.groupby('category').conservation_status.count()
print(protected_animals)

conservation_statuses = species.groupby('conservation_status').category.count()
print(conservation_statuses)
```

```python
f, ax = plt.subplots(figsize=(8, 6))
plt.bar(range(len(protected_animals)), protected_animals)
ax.set_xticks([0, 1, 2, 3, 4, 5, 6])
ax.set_xticklabels(['Amphibian', 'Bird', 'Fish', 'Mammal', 'Nonvascular Plant', 'Reptile', 'Vascular Plant'], rotation=30)
plt.title('Number of Protected Species in Each Category')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
plt.bar(range(len(conservation_statuses)), conservation_statuses)
ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(['Endangered', 'In Recovery', 'Species of Concern', 'Threatened'], rotation=30)
plt.title('Number of Species in Each Conservation Status Category')
plt.show()
```
